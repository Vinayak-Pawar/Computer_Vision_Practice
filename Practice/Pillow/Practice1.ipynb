```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grayscale image (8x8) with intensity values from 0 to 255
gray_image = np.array([
    [0, 32, 64, 96, 128, 160, 192, 224],
    [0, 32, 64, 96, 128, 160, 192, 224],
    [0, 32, 64, 96, 128, 160, 192, 224],
    [0, 32, 64, 96, 128, 160, 192, 224],
    [0, 32, 64, 96, 128, 160, 192, 224],
    [0, 32, 64, 96, 128, 160, 192, 224],
    [0, 32, 64, 96, 128, 160, 192, 224],
    [0, 32, 64, 96, 128, 160, 192, 224]
], dtype=np.uint8)

# Display the grayscale image
plt.figure(figsize=(5, 5))
plt.imshow(gray_image, cmap='gray', vmin=0, vmax=255)
plt.colorbar()
plt.title("Grayscale Image with Intensities from 0 to 255")
plt.show()
```

```python
# Create a color image (3x3) with distinct RGB channels
color_image = np.zeros((3, 3, 3), dtype=np.uint8)

# Red channel
color_image[0, :, 0] = [255, 0, 0]  # Red
# Green channel
color_image[1, :, 1] = [0, 255, 0]  # Green
# Blue channel
color_image[2, :, 2] = [0, 0, 255]  # Blue

# Display the color image
plt.figure(figsize=(5, 5))
plt.imshow(color_image)
plt.title("Color Image with RGB Channels")
plt.show()
```

```python
import cv2

# Create an example image (5x5) with intensity values set to 127
image = np.ones((5, 5, 3), dtype=np.uint8) * 127  # Example image

# Find the maximum and minimum intensity values
max_intensity = image.max()
min_intensity = image.min()

print("Max intensity:", max_intensity)
print("Min intensity:", min_intensity)

# Display the image
plt.figure(figsize=(5, 5))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Example Image")
plt.show()
```
